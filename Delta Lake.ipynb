### Managed Table

```python
%sql
create table salesdb.mantable(
  id int,
  name string,
  marks int
)
using delta
```

```python
%sql
insert into salesdb.mantable values(1,'a',10),(2,'b',20),(3,'c',30),(4,'d',40)
 
```

```python
%sql
select * from salesdb.mantable
```

```python
%sql
drop table salesdb.mantable
```

### External Table

```python
%sql
    
create table salesdb.exttable(
  id int,
  name string,
  marks int
)
using delta
location 'abfss://destination@learningdb.dfs.core.windows.net/salesdb/exttable'
```
